# Ajustemos algunos modelos

Por el momento no entraremos en los detalles de los hiperpar谩metros de cada modelo, ni separaremos el conjunto de datos en train set y test set.

Para ajustar algunos modelos, trabajaremos tambien con el breast cancer wisconsin dataset, un famoso conjunto de datos para clasificaci贸n binaria, no entraremos en detalles, pero si gustas puedes revisarlos [aqu铆](https://archive.ics.uci.edu/dataset/17/breast+cancer+wisconsin+diagnostic).

```python
data = datasets.load_breast_cancer()
```

```python
print(data.DESCR)
```

```python
Xbc = pd.DataFrame(data.data, columns=data.feature_names)

ybc = data.target

Xbc.head()
```

```python
Xbc = Xbc.iloc[:, :10] # Trabajaremos con los valores promedios de las caracteristicas

Xbc.head()
```

Ajustemos los modelos que han visto hasta el momento. Partiremos por una regresi贸n log铆stica que tiene dos implementaciones conocidas en python.

## Regresi贸n Log铆stica

La implementaci贸n de regresi贸n log铆stica de `Statsmodels`, es m谩s completa que la de `scikitlearn`, ya que adem谩s de ajustar el modelo, permite calcular los test estad铆sticos y an谩lisis necesarios para verificar que se cumplen las condiciones sobre las que se basa este tipo de modelos.

Agregamos una columna de 1's para a帽adir intercepto al modelo.

```python
import statsmodels.api as sm

sm.add_constant(Xbc)
```

```python
logit_model = sm.Logit(ybc, sm.add_constant(Xbc)).fit()

print(logit_model.summary())
```

Si analizamos las estad铆sticas que entrega el modelo, podr铆amos descatar algunas covariables. Supongamos queremos descartar la textura y la concavidad (solo es un ejemplo) del modelo, para ajustar uno nuevo.

```python
Xbc2 = Xbc.drop(['mean texture', 'mean concavity'], axis = 1)
Xbc2.head()
```

```python
logit_model2 = sm.Logit(ybc, sm.add_constant(Xbc2)).fit()

print(logit_model2.summary())
```

Predigamos alg煤na observaci贸n usando el primer modelo ajustado.

```python
test = Xbc.iloc[2:5]

pred = logit_model.predict(sm.add_constant(test)) # recuerda que nos entrega la probabilidad de pertenecer a la clase 1, en nuestro caso Benign.
pred
```

```python
np.where(pred > 0.5, 1, 0)
```

Veamos la implementaci贸n de `sklearn`.

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression(random_state=0, max_iter = 1000).fit(Xbc, ybc)
```

Predigamos una observaci贸n:

```python
lr.predict(test)
```

Veamos la probabilidad de pertenecer a cada clase.

```python
lr.predict_proba(test)
```

## Naive Bayes

Para ajustar el modelo Naive Bayes, debemos tener sospecha de que las covariables son independientes, para eso observamos la correlaci贸n entre estas.

En adelante trabajaremos con el `Iris` dataset.

```python
df.corr(numeric_only=True)
```

```python
mask = np.tril(np.ones_like(df.corr(numeric_only=True), dtype=bool), k = -1)

sns.heatmap(df.corr(numeric_only=True), annot = True, vmin = -1, vmax = 1, cmap = 'coolwarm', mask = mask);
```

La interpretaci贸n de resultados es un paso clave al momento de ajustar modelos. Supongamos que se puede ajustar un Naive Bayes ;)

```python
from sklearn.naive_bayes import GaussianNB
```

```python
X = df.drop(['species'], axis = 1)
y = df['species']
```

```python
nb = GaussianNB().fit(X,y)
```

```python
test = pd.DataFrame({
    'sepal length (cm)': [3],
    'sepal width (cm)': [3],
    'petal length (cm)': [2],
    'petal width (cm)': [2]
})
test
```

```python
nb.predict(test)
```

```python
nb.predict_proba(test)
```
