# Ajustemos algunos modelos

Por el momento no entraremos en los detalles de los hiperparámetros de cada modelo, ni separaremos el conjunto de datos en train set y test set.

Para ajustar algunos modelos, trabajaremos tambien con el breast cancer wisconsin dataset, un famoso conjunto de datos para clasificación binaria, no entraremos en detalles, pero si gustas puedes revisarlos [aquí](https://archive.ics.uci.edu/dataset/17/breast+cancer+wisconsin+diagnostic).

```python
data = datasets.load_breast_cancer()
```

```python
print(data.DESCR)
```

```python
Xbc = pd.DataFrame(data.data, columns=data.feature_names)

ybc = data.target

Xbc.head()
```

```python
Xbc = Xbc.iloc[:, :10] # Trabajaremos con los valores promedios de las caracteristicas

Xbc.head()
```

Ajustemos los modelos que han visto hasta el momento. Partiremos por una regresión logística que tiene dos implementaciones conocidas en python.

## Regresión Logística

La implementación de regresión logística de `Statsmodels`, es más completa que la de `scikitlearn`, ya que además de ajustar el modelo, permite calcular los test estadísticos y análisis necesarios para verificar que se cumplen las condiciones sobre las que se basa este tipo de modelos.

Agregamos una columna de 1's para añadir intercepto al modelo.

```python
import statsmodels.api as sm

sm.add_constant(Xbc)
```

```python
logit_model = sm.Logit(ybc, sm.add_constant(Xbc)).fit()

print(logit_model.summary())
```

Si analizamos las estadísticas que entrega el modelo, podríamos descatar algunas covariables. Supongamos queremos descartar la textura y la concavidad (solo es un ejemplo) del modelo, para ajustar uno nuevo.

```python
Xbc2 = Xbc.drop(['mean texture', 'mean concavity'], axis = 1)
Xbc2.head()
```

```python
logit_model2 = sm.Logit(ybc, sm.add_constant(Xbc2)).fit()

print(logit_model2.summary())
```

Predigamos algúna observación usando el primer modelo ajustado.

```python
test = Xbc.iloc[2:5]

pred = logit_model.predict(sm.add_constant(test)) # recuerda que nos entrega la probabilidad de pertenecer a la clase 1, en nuestro caso Benign.
pred
```

```python
np.where(pred > 0.5, 1, 0)
```

Veamos la implementación de `sklearn`.

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression(random_state=0, max_iter = 1000).fit(Xbc, ybc)
```

Predigamos una observación:

```python
lr.predict(test)
```

Veamos la probabilidad de pertenecer a cada clase.

```python
lr.predict_proba(test)
```

## Naive Bayes

Para ajustar el modelo Naive Bayes, debemos tener sospecha de que las covariables son independientes, para eso observamos la correlación entre estas.

En adelante trabajaremos con el `Iris` dataset.

```python
df.corr(numeric_only=True)
```

```python
mask = np.tril(np.ones_like(df.corr(numeric_only=True), dtype=bool), k = -1)

sns.heatmap(df.corr(numeric_only=True), annot = True, vmin = -1, vmax = 1, cmap = 'coolwarm', mask = mask);
```

La interpretación de resultados es un paso clave al momento de ajustar modelos. Supongamos que se puede ajustar un Naive Bayes ;)

```python
from sklearn.naive_bayes import GaussianNB
```

```python
X = df.drop(['species'], axis = 1)
y = df['species']
```

```python
nb = GaussianNB().fit(X,y)
```

```python
test = pd.DataFrame({
    'sepal length (cm)': [3],
    'sepal width (cm)': [3],
    'petal length (cm)': [2],
    'petal width (cm)': [2]
})
test
```

```python
nb.predict(test)
```

```python
nb.predict_proba(test)
```

## Árbol de Clasificación

Ajustemos un Árbol de Clasificación.

```python
from sklearn.tree import DecisionTreeClassifier, plot_tree
```

```python
dt = DecisionTreeClassifier(criterion = 'gini', max_depth = 5, min_samples_split = 4).fit(X,y)
```

Visualicemos el árbol obtenido.

```python
plt.figure(figsize = (10,8))
plot_tree(dt, feature_names = iris.feature_names)
plt.show()
```

```python
dt.predict(test)
```

```python
dt.predict_proba(test)
```
