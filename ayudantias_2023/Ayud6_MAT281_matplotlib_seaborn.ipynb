# Matplotlib

Matplotlib es una de las bibliotecas de visualización de datos más utilizadas y poderosas en el ecosistema de Python. Proporciona una amplia variedad de opciones para crear gráficos estáticos, interactivos y personalizables. Algunas de las características destacadas de Matplotlib son:

+ Versatilidad: Matplotlib es altamente versátil y permite crear una amplia gama de gráficos, incluyendo gráficos de línea, barras, dispersión, histogramas, gráficos de pastel y más.

+ Control detallado: Ofrece un control detallado sobre cada elemento del gráfico, como colores, etiquetas, títulos, ejes y leyendas, lo que permite personalizar la apariencia de la visualización según las necesidades específicas.

```python
import matplotlib.pyplot as plt

import numpy as np
```

## Mi primer gráfico

Partamos por los gráficos básicos ejemplificando parámetros, ya que la mayoría de gráficos comparten estos parámetros.

Para definir el tamaño de la figura creamos un entorno `plt.figure()`

```python
# Creamos datos de ejemplo
x = np.linspace(-4,4,1_000)
y = np.sin(x)

plt.figure(
    figsize=(8,2) # (Ancho, Alto) del entorno figura
)

plt.plot(x,y) # Graficar

plt.show() # Mostrar el gráfico
```

## Etiquetas

Mejoremos el gráfico agregando etiquetas al Título, Ejes y Leyendas.

```python
plt.figure(figsize=(8,3))
plt.plot(x,y,
         label = 'Ejemplo' # Nombre distintivo de un gráfico       
        )

plt.xlabel('Eje X', loc='right', rotation = 45)
plt.ylabel('Eje Y', loc='top', alpha = 0.5)
plt.title('Mi primer gráfico', loc='left', fontsize = 15,)

plt.legend(
    #loc = 'best' # Por defecto busca la mejor ubicación
    loc = 'upper right', # 'center', 'lower left'
    title  = 'Etiquetas',
    shadow = True
) 

plt.show()
```

Para más detalles del formato de texto en gráficos visita la [documentación](https://matplotlib.org/stable/api/text_api.html#matplotlib.text.Text).

## Personalizar gráfico

Además, podemos personalizar el gráfico, como tipo de linea, color, ancho, etc. A continuación, te presento alguna de los parámetros que ofrece el paquete.

```python
x = [1, 2, 3, 4, 5]
y = [2, 4, 1, 3, -1]

plt.plot(x,y,
         color = 'm',
         ls = ':',     # linestyle = '-', '--', '-.', ':', 'None', ' ', ''
         lw = 2,       # linewidth = 0.2
         marker = 's', # 'o', '8', '*', '1' etc
         ms = 10       # markersize = 5
         # ...
        )
plt.xlim(-2, 10) # left, right
plt.ylim(-3, 8) # bottom, top 

# Generar cuadrídula
plt.grid(ls='-.',alpha=0.5,lw=2,
         axis = 'x' # 'both' (por defecto), 'x', 'y'
         )
```

Para más detalles revisa la [documentación](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html).

## Más de un gráfico

En una misma figura puedes realizar más de un gráfico.

```python
y2 = [2, 4, 4, 6, 5]
y3 = [4, -1, 0, 6, 5]
plt.figure(figsize = (8,4))
plt.plot(x,y, lw = 12)
plt.plot(x,y2,lw = 6, c ='r')
plt.plot(x,y3,lw = 3, c ='k')

plt.title('Más de un gráfico'); plt.legend(['Línea 1', 'Línea 2', 'Línea 3']);
```

No hay mucho que decir, puedes combinar los gráficos como quieras. Sin embargo, hay un parámetro no muy conocido que te permite ordenar la superposición de los gráficos: `zorder`. Nota que por defecto queda por debajo el primer gráfico que se define.

+ zorder: Set the zorder for the artist. Artists with lower zorder values are drawn first.

```python
plt.figure(figsize = (8,4))
plt.plot(x,y,lw = 12, zorder = 3)
plt.plot(x,y2,lw = 6, c ='r', zorder = 2)
plt.plot(x,y3,lw = 3, c ='k', zorder = 1)

plt.title('Más de un gráfico'); plt.legend(['Línea 1', 'Línea 2', 'Línea 3']);
```

## [Gráficos de dispersión](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.scatter.html)

Una herramienta esencial en el análisis de datos son **los gráficos de puntos** o **scatterplots**. Estas representaciones gráficas nos permiten visualizar relaciones y patrones entre dos variables, lo que es crucial para comprender la correlación y la dispersión en nuestros datos.

| Aspecto | Scatter Plots |
| ------- | ------------- |
| **Tipo de Datos** | Numéricos |
| **Visualización** | Relaciones entre variables |
| **Descubrimiento** | Identificación de patrones y correlaciones |
| **Toma de Decisiones** | Evaluación de relaciones y tendencias |

### Explorando Relaciones

Este tipo de gráficos nos permiten ver si dos variables están relacionadas de alguna manera y cómo se dispersan los puntos de datos en el gráfico. Imagina poder visualizar la relación entre la edad y los ingresos de una población o la correlación entre el tiempo de estudio y las calificaciones de los estudiantes. 

### Importancia en la Identificación de Patrones y Correlaciones

Uno de los aspectos más destacados de los scatter plots es su capacidad para identificar patrones y correlaciones. Al observar la dispersión de los puntos en el gráfico, podemos detectar tendencias, relaciones positivas o negativas, y puntos de datos atípicos que pueden tener un impacto significativo en nuestro análisis.

Un ejemplo recurrente es comparar resultados de predicción en un problema de regresión.

```python
# Datos de ejemplo
x = np.linspace(0,100,1_000)
y = x + np.random.normal(0,5,1_000)

plt.scatter(x,y,c='w', 
            marker = 'o',     # '.', 'v', '+', '_', '|' 
            s = 50,           # Size de los puntos
            edgecolors = 'g', # Color del borde
            # ...
            )

plt.plot([0,100], [0,100], c = 'k')
plt.title('Predicción vs Real'); plt.xlabel('Real'); plt.ylabel('Predicción');
```

## Gráfico de barras e Histograma

Continuemos con dos herramientas fundamentales al analizar datos: los **gráficos de barra** y los **histogramas**.

A primera vista podrían parecer iguales, pero es importante conocer la diferencia.

| Aspecto | Gráficos de Barra | Histogramas |
| ------- | ----------------- | ----------- |
| **Tipo de Datos** | Categóricos/Discretos | Numéricos |
| **Visualización** | Comparación de categorías | Distribución de datos |
| **Toma de Decisiones** | Comunicación efectiva | Análisis de tendencias |

### Gráficos de barras [[1]](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.bar.html) [[2]](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.barh.html#matplotlib.pyplot.barh)

```python
# Datos de ejemplo
x = ['Perro', 'Gato', 'Pájaro', 'Oso']
y = [5, 6, 7, 1]

plt.bar(x,y,color='m',edgecolor='k',
        width = 0.5,   # Ancho de barras
        align = 'edge' # Alineación (por defecto: 'center')
        # Combinando 'edge' con width negativo se alinea a la izq.
        # ...
        )
plt.title('Animales');
```

```python
plt.barh(x,y,color='m',edgecolor='k',
         height = 0.5,  # Ancho de barras
         align = 'edge' # Alineación (por defecto: 'center')
         # Combinando 'edge' con height negativo se alinea hacia abajo del centro.
         # ...
        )
plt.title('Animales');
```

### [Histograma](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.hist.html)

```python
edades = np.random.normal(30, 10, 1_000)
plt.hist(edades,edgecolor='k',color='m',
         bins = 15,          # Cantidad de barras
         density = False,    # Normalización
         cumulative = False, # Histograma acumulativo
         orientation = 'vertical',
         # ...
         )
plt.title('Histograma de Edades');
```

## [Boxplot](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.boxplot.html)

Otro tipo de gráfico útil para comprender la distribución y la variabilidad de nuestros datos.

| Aspecto | Boxplots |
| ------- | -------- |
| **Tipo de Datos** | Numéricos |
| **Visualización** | Resumen de distribuciones |
| **Descubrimiento** | Identificación de valores atípicos |
| **Toma de Decisiones** | Evaluación de distribuciones |

### Importancia en la Detección de Valores Atípicos

Una de las características más destacadas de los boxplots es su capacidad para detectar valores atípicos o extremos en los datos. Los valores que se encuentran significativamente fuera de los "bigotes" del boxplot se consideran atípicos y pueden ser señales importantes de problemas o patrones inesperados en los datos.


```python
plt.boxplot(edades,
            widths = 0.5, # Ancho del box
            vert = True,  # Orientación del boxplot
            sym = 'r*',   # Controla la apariencia de los outliers '[color][marker]'
            labels = ['Edades'], # Etiquetas
            boxprops={'c':'b', 'lw':3, 'alpha':0.5}, # Estilo del box
            medianprops={'c':'b', 'ls':':'},         # Estilo de la mediana
            whiskerprops={'lw':1, 'ls':'-.'},        # Estilo de los 'bigotes'
            capprops={'lw':2, 'c':'b'},              # Estilo de las 'tapas'
            );
```

## Varios gráficos en el mismo entorno

Hasta ahora, hemos explorado herramientas visuales clave en la ciencia de datos, como gráficos de barra, histogramas, boxplots y scatterplots, presentando un solo gráfico en cada figura. Sin embargo, podemos crear múltiples gráficos en un solo lienzo. ¡Bienvenidos a los **subplots**!


Los subplots nos permiten organizar múltiples gráficos en una sola figura. Esto es esencial cuando queremos comparar diferentes conjuntos de datos o representar múltiples aspectos de nuestros datos en una narrativa visual. Imagine poder comparar la distribución de ventas en diferentes regiones en un solo lienzo.


Para dar el siguiente paso en nuestro viaje de visualización, aprenderemos a crear subplots. Esto implica dividir una figura en una cuadrícula de subgráficos y colocar visualizaciones en cada uno de ellos.

```python
# Definir una figura y una cuadrícula de subgráficos (1 filas, 2 columnas)
fig, axs = plt.subplots(1, 2, figsize=(8, 4),
                        sharex=False, # Compartir ejes x
                        sharey=False,  # Compartir ejes y 
                        )

# Subplot 1
axs[0].plot([1, 2, 3, 4], [1, 4, 9, 16], ls='-.')
axs[0].plot([1, 2, 3, 4], [1, 5, 7, 10], lw=2, marker='*')

# Subplot 2
axs[1].bar(['A', 'B', 'C'], [3, 7, 2], color=['r','g','m'], edgecolor='k')

# Ajustar el espaciado entre subplots
plt.tight_layout();
```

Los parámetros internos de cada tipo de gráfico que hemos visto se mantienen.

En cambio, la algunos métodos (no todos) como `title()`, `xlabel()`, `ylabel()`, `xlim`, `ylim`, etc. se les antepone un `set_`. Otros como `legend()` o `grid()` se mantienen.

```python
fig, axs = plt.subplots(1, 2, figsize=(8, 4))

# Subplot 1
axs[0].plot([1, 2, 3, 4], [1, 4, 9, 16], ls='-.', label='Línea 1')
axs[0].plot([1, 2, 3, 4], [1, 5, 7, 10], lw=2, marker='*', label='Línea 1')
axs[0].legend()
axs[0].set_xlabel('Eje X')
axs[0].set_ylabel('Eje Y')
axs[0].set_xlim(0,5)

# Subplot 2
axs[1].bar(['A', 'B', 'C'], [3, 7, 2], color=['r','g','m'], edgecolor='k', zorder = 2)
axs[1].grid(axis = 'y', zorder = 1)
axs[1].set_title('Ejemplo 2')

# Ajustar el espaciado entre subplots
plt.tight_layout();
```

# Seaborn

Seaborn es una poderosa extensión de Matplotlib que nos permite llevar nuestras visualizaciones de datos al siguiente nivel. Ofrece una interfaz de alto nivel para crear gráficos atractivos y una serie de ventajas que hacen que sea una elección popular en el mundo de la ciencia de datos.

### Ventajas de Utilizar Seaborn
+ Estilo Atractivo: Seaborn viene con estilos y paletas de colores predefinidos que hacen que sus gráficos sean más atractivos y estéticamente agradables.

+ Sintaxis Simplificada: Seaborn simplifica la creación de gráficos complejos. Con solo unas pocas líneas de código, puedes crear visualizaciones avanzadas.

+ Integración con Pandas: Seaborn se integra perfectamente con Pandas DataFrames, lo que facilita la visualización de datos almacenados en estructuras de datos de Pandas.

+ Visualización Estadística: Seaborn se centra en la visualización de datos estadísticos y proporciona herramientas para visualizar relaciones y distribuciones de manera efectiva.

```python
import seaborn as sns

import pandas as pd
```

Motivaremos algunos ejemplos con el dataset `tips`.

```python
data = sns.load_dataset("tips")
data.head()
```

No entraremos en tantos detalles, sino que veremos la flexibilidad que nos entrega este paquete. Por ejemplo, podemos hacer depender de una variable características del gráfico. En el caso de un [`scatterplot()`](https://seaborn.pydata.org/generated/seaborn.scatterplot.html) podemos depende del tamaño de los puntos, color, marcador, etc. 

```python
# Gráfico de dispersión de cuenta total vs. propina
sns.scatterplot(x="total_bill", y="tip", data=data,
                hue = 'sex',
                style = 'time',
                size = 'time'
                )

plt.title("Relación entre la cuenta total y la propina")
plt.xlabel("Cuenta total ($)");plt.ylabel("Propina ($)");
```

Si lo queremos visualizar por separado podemos usar un [`relplot()`](https://seaborn.pydata.org/generated/seaborn.relplot.html).

```python
sns.relplot(x="total_bill", y="tip", data=data,
            col='time', hue="sex", style="sex",
            kind="scatter"
            );
```

Otro ejemplo de gráfico es un [`boxplot()`](https://seaborn.pydata.org/generated/seaborn.boxplot.html).

```python
sns.boxplot(x="day", y="total_bill", data=data, hue='sex')

plt.title("Distribución de la cuenta total por día y sexo")
plt.xlabel("Día de la semana");plt.ylabel("Cuenta total ($)");
```

Como último ejemplo, si quieres hacer las comparativas entre columnas de un dataframe, tenemos el tipo de gráfico [`pairplot()`](https://seaborn.pydata.org/generated/seaborn.pairplot.html).

```python
iris = sns.load_dataset("iris")
sns.pairplot(iris, hue="species", palette="Set2", diag_kind='hist', height=2.5, corner = True);
```

Como bonus te presento la función [`PairGrid()`](https://seaborn.pydata.org/generated/seaborn.PairGrid.html), revisa su documentación!

```python
g = sns.PairGrid(iris, hue='species')
g.map_upper(sns.scatterplot)
g.map_lower(sns.kdeplot)
g.map_diag(sns.histplot, multiple="stack")
g.add_legend();
```

##### 2.1 Visualiza la proporción de pasajeros que sobrevivieron frente a los que no sobrevivieron, usa una [paleta de colores](https://seaborn.pydata.org/tutorial/color_palettes.html#qualitative-color-palettes) de tu gusto.

```python
with sns.color_palette('pastel'):
    sns.countplot(data=titanic, x='survived', edgecolor='k')
    plt.title('Supervivencia en el Titanic')
    plt.xlabel('Sobrevivientes'); plt.ylabel('Número de Pasajeros');
```

#### 2.2 En un mismo entorno visualiza para las variables categóricas la cantidad de pasajeros según la supervivencia (`survived`). 

```python
titanic.select_dtypes(include='object').columns
```

```python
# Escribe tu code aquí

fig, axs = plt.subplots(2,2,figsize=(12,12),sharey=True)

for i, column in enumerate(titanic.select_dtypes(include='object').columns):
    sns.countplot(data=titanic, x=column, hue="survived", ax=axs[i//2, i%2], edgecolor='k')
    axs[i//2, i%2].set_title(f'Supervivencia por {column}')
    axs[i//2, i%2].set_xlabel(f'{column}')
    axs[i//2, i%2].set_ylabel(f'Número de Pasajeros')
    axs[i//2, i%2].legend(title="Supervivencia", labels=["No Sobrevivió", "Sobrevivió"])
    
plt.tight_layout();
```
