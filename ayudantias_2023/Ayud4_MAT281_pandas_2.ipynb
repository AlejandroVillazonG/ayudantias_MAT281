### Concatenar

Imagina que tienes varias tablas con las mismas columnas y quieres unirlas en una grande. La función `concat` está diseñada para esta labor.

![concat](https://pandas.pydata.org/pandas-docs/stable/_images/merging_concat_basic.png)

````python
df1 = pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3'],
                    'B': ['B0', 'B1', 'B2', 'B3'],
                    'C': ['C0', 'C1', 'C2', 'C3'],
                    'D': ['D0', 'D1', 'D2', 'D3']},
                   index=[0, 1, 2, 3])


df2 = pd.DataFrame({'A': ['A4', 'A5', 'A6', 'A7'],
                    'B': ['B4', 'B5', 'B6', 'B7'],
                    'C': ['C4', 'C5', 'C6', 'C7'],
                    'D': ['D4', 'D5', 'D6', 'D7']},
                   index=[4, 5, 6, 7])


df3 = pd.DataFrame({'A': ['A8', 'A9', 'A10', 'A11'],
                    'B': ['B8', 'B9', 'B10', 'B11'],
                    'C': ['C8', 'C9', 'C10', 'C11'],
                    'D': ['D8', 'D9', 'D10', 'D11']},
                   index=[8, 9, 10, 11])

pd.concat([df1, df2, df3])
````

Si los dataframes tienen distintas columnas no es impedimento para concatenarlas.

````python
df4 = pd.DataFrame({'B': ['B2', 'B3', 'B6', 'B7'],
                    'D': ['D2', 'D3', 'D6', 'D7'],
                    'F': ['F2', 'F3', 'F6', 'F7']},
                    index=[2, 3, 6, 7])

pd.concat([df1, df4], ignore_index= True)
````

¿Quieres que se agreguen columnas dependiendo de los index que tienen cada dataframe?

![concat3](https://pandas.pydata.org/pandas-docs/stable/_images/merging_concat_axis1.png)

````python
pd.concat([df1, df4], axis=1)
````

Por defecto la función `.concat()` concatena a lo largo del axis=0.

### Unir

En pandas es posible unir dos tablas sin la necesidad de iterar fila por fila. La funcionalidad la entrega el método `merge`.

Existen distintos tipos de unir datos. Los cuatro tipos de cruces más comunes:

- ```inner```: (_default_) Retorna aquellos registros donde los valores de columnas utilizadas para los cruces se encuentran en ambas tablas.
- ```left```: Retorna todos los registros de la tabla colocada a la izquierda, aunque no tengan correspondencia en la tabla de la derecha.
- ```right```: Retorna todos los registros de la tabla colocada a la derecha, aunque no tengan correspondencia en la tabla de la izquierda.
- ```outer```: Retorna todos los valores de ambas tablas, tengan correspondencia o no.

La siguiente imagen explica el resultado que se obtiene con los distintos tipos de cruces.

![Joins](https://dataschool.com/assets/images/how-to-teach-people-sql/sqlJoins/sqlJoins_7.png?w=662&h=361)
![Joins2](https://lukaseder.files.wordpress.com/2016/07/venn-join1.png?w=662&h=361)

Veamos otros _ejemplitos_

````python
left = pd.DataFrame({'key1': ['K0', 'K0', 'K1', 'K2'],
                     'key2': ['K0', 'K1', 'K0', 'K1'],
                     'A': ['A0', 'A1', 'A2', 'A3'],
                     'B': ['B0', 'B1', 'B2', 'B3']})


right = pd.DataFrame({'key1': ['K0', 'K1', 'K1', 'K2'],
                      'key2': ['K0', 'K0', 'K0', 'K0'],
                      'C': ['C0', 'C1', 'C2', 'C3'],
                      'D': ['D0', 'D1', 'D2', 'D3']})

display(left)

display(right)
````

__Left Merge__

````python
pd.merge(left, right, how = 'left', on=['key1', 'key2'])
````

__Right Merge__

![right_merge](https://pandas.pydata.org/pandas-docs/stable/_images/merging_merge_on_key_right.png)

````python
pd.merge(left, right, how='right', on=['key1', 'key2'])
````

__Outer Merge__

![outer_merge](https://pandas.pydata.org/pandas-docs/stable/_images/merging_merge_on_key_outer.png)

````python
pd.merge(left, right, how='outer', on=['key1', 'key2'])
````

__Inner Merge__

![inner_merge](https://pandas.pydata.org/pandas-docs/stable/_images/merging_merge_on_key_inner.png)

````python
pd.merge(left, right, on=['key1', 'key2'])
````

__Problemas de llaves duplicadas__

Cuando se quiere realizar el cruce de dos tablas, pero en ambas tablas existe una columna (key) con el mismo nombre, para diferenciar la información entre la columna de una tabla y otra, pandas devuelve el nombre de la columna con un guión bajo x (key_x) y otra con un guión bajo y (key_y)

````python
left2 = pd.DataFrame({'A': [1, 2], 'B': [2, 2]})
right2 = pd.DataFrame({'A': [4, 5, 6], 'B': [2, 2, 2]})

display(left2)
display(right2)

pd.merge(left2, right2, on='B', how='outer')
````

Tambien podemos cruzar todos los datos de ambas tablas.

````python
left.merge(right, how = 'cross', suffixes= ['_left', '_right'])
````

No es necesario que la columna compartida tenga el mismo nombre!

````python
left.merge(right, how = 'right', left_on = 'key2', right_on = 'key1', suffixes= ['_left', '_right'])
````

Considera que `merge` también se puede usar como método, por lo que es posible concatener varias operaciones.
